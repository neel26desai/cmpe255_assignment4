Identifying airlines having most delays

```python
  # Top carriers with the most delays (number of flights delayed more than 15 minutes)
top_carriers_delays = airline_delay.groupby('carrier_name')['arr_del15'].sum().sort_values(ascending=False).head(10)

top_carriers_delays
```

```python
# Convert the top carriers with the most delays data to a format suitable for D3.js
top_carriers_list = top_carriers_delays.reset_index().rename(columns={'index': 'carrier', 'arr_del15': 'delays'}).to_dict(orient='records')
# Convert the list of dictionaries to a JavaScript array of objects (as a string)
data_string = str(top_carriers_list).replace("'", '"')
print(data_string)
# D3.js Interactive Bar Graph with Dynamically Injected Data
html_code = f"""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <script src="https://d3js.org/d3.v7.min.js"></script>
    <style>
        .bar {{
            fill: steelblue;
        }}
        .bar:hover {{
            fill: orange;
        }}
        .axis-label {{
            font-size: 14px;
        }}
    </style>
</head>
<body>
<script>
    // Injected data for the bar graph
    var data = {data_string};

    // SVG dimensions
    var margin = {{top: 20, right: 30, bottom: 70, left: 150}},
        width = 800 - margin.left - margin.right,
        height = 500 - margin.top - margin.bottom;

    // X and Y scales
    var x = d3.scaleLinear().range([0, width]);
    var y = d3.scaleBand().range([height, 0]).padding(0.1);

    // Create SVG element
    var svg = d3.select("body").append("svg")
        .attr("width", width + margin.left + margin.right)
        .attr("height", height + margin.top + margin.bottom)
        .append("g")
        .attr("transform", "translate(" + margin.left + "," + margin.top + ")");

    // Tooltip div
    var tooltip = d3.select("body").append("div")
        .attr("class", "tooltip")
        .style("position", "absolute")
        .style("z-index", "10")
        .style("visibility", "hidden")
        .style("background", "#F9F9F9")
        .style("padding", "8px")
        .style("border-radius", "6px")
        .style("border", "1px solid #ccc")
        .style("box-shadow", "0px 0px 4px #aaa");

    x.domain([0, d3.max(data, function(d) {{ return d.delays; }})]);
    y.domain(data.map(function(d) {{ return d.carrier_name; }}));

    // Add bars
    svg.selectAll(".bar")
        .data(data)
        .enter().append("rect")
        .attr("class", "bar")
        .attr("y", function(d) {{ return y(d.carrier_name); }})
        .attr("height", y.bandwidth())
        .attr("x", 0)
        .attr("width", function(d) {{ return x(d.delays); }})
        .on("mouseover", function(event, d) {{
            d3.select(this).attr("fill", "#FFD700");
            var coords = d3.pointer(event);
            tooltip.html(d.carrier_name + ": " + d.delays + " delays")
                .style("left", (coords[0] + 150) + "px")
                .style("top", (coords[1] - 30) + "px");
            tooltip.style("visibility", "visible");
        }})
        .on("mouseout", function() {{
            d3.select(this).attr("fill", "steelblue");
            tooltip.style("visibility", "hidden");
        }});

    // Add X and Y axes
    svg.append("g")
        .attr("transform", "translate(0," + height + ")")
        .call(d3.axisBottom(x))
        .append("text")
        .attr("y", 30)
        .attr("x", width / 2)
        .attr("text-anchor", "middle")
        .attr("class", "axis-label")
        .text("Number of Delays");

    svg.append("g")
        .call(d3.axisLeft(y))
        .append("text")
        .attr("transform", "rotate(-90)")
        .attr("y", -120)
        .attr("x", -height / 2)
        .attr("text-anchor", "middle")
        .attr("class", "axis-label")
        .text("Carrier");
</script>
</body>
</html>
"""

# Display in Colab
display(HTML(html_code))


```

```python
# D3.js code for the pie chart
d3_pie_script = f"""
<script src="https://d3js.org/d3.v5.min.js"></script>
<div id="d3-pie-container"></div>
<script>
    var data = {d};
    var width = 400;
    var height = 400;
    var radius = Math.min(width, height) / 2;

    var color = d3.scaleOrdinal(d3.schemeCategory10);

    var pie = d3.pie().value(d => d.value);
    var arc = d3.arc().outerRadius(radius - 10).innerRadius(0);

    var svg = d3.select("#d3-pie-container")
                .append("svg")
                .attr("width", width)
                .attr("height", height)
                .append("g")
                .attr("transform", "translate(" + width / 2 + "," + height / 2 + ")");

    var g = svg.selectAll(".arc")
               .data(pie(d3.entries(data)))
               .enter().append("g")
               .attr("class", "arc");

    g.append("path")
     .attr("d", arc)
     .style("fill", d => color(d.data.key))
     .on("mouseover", function(d) {{
         d3.select(this)
           .style("fill", d3.rgb(color(d.data.key)).darker(2));
         svg.append("text")
            .attr("class", "label")
            .attr("text-anchor", "middle")
            .attr("y", 5)
            .text(d.data.key + ": " + d.data.value);
     }})
     .on("mouseout", function(d) {{
         d3.select(this)
           .style("fill", color(d.data.key));
         d3.selectAll(".label").remove();
     }});

    g.append("text")
     .attr("transform", d => "translate(" + arc.centroid(d) + ")")
     .attr("dy", ".35em")
     .style("text-anchor", "middle")
     .text(d => d.data.key);
</script>
"""

# Display the D3.js pie chart
display(HTML(d3_pie_script))
```

Based on the exploratory data analysis (EDA) we've conducted on the airline delay dataset for December 2019 and 2020, here are some conclusions and insights:

General Observations:

The dataset contains information on delays for different airlines across various airports for December of 2019 and 2020.
Most of the variables in the dataset are numerical, representing counts and durations of different types of delays.
Data Quality:

The dataset had missing values, especially in columns related to delay counts and durations. This could mean that there were no delays or the data might not have been recorded.
No apparent outliers were detected using the box plot visualization, suggesting the data is relatively consistent in its range.
Airports with Most Delays:

"Hartsfield-Jackson Atlanta International" had the highest number of delays followed by airports like "Dallas/Fort Worth International" and "Chicago O'Hare International".
It's essential to note that these are major airports with high traffic, which might be a contributing factor to the high number of delays.
Reasons for Delays:

The most common reasons for delays were carrier_ct (delays due to the air carrier) and late_aircraft_ct (delays because a previous flight on the same airplane was late).
Delays due to security (security_ct) were the least common.
This suggests that internal airline logistics and previous flight delays have a more significant impact on flight punctuality than external factors like security or even weather.
Yearly Comparison:

Our analysis was restricted to the month of December for both 2019 and 2020. However, it would be interesting to see if there's a significant difference in delays between these years, especially given the impact of the COVID-19 pandemic on air travel in 2020.
Distribution of Delays:

Most of the airports had a similar distribution of delay reasons, with carrier and late aircraft being the primary reasons across the board.
Potential Recommendations:

Airlines could focus on improving internal logistics and ensuring punctuality for each flight to reduce the cascading effect of one delayed flight causing subsequent delays.
Airports with a high number of delays, like Atlanta and Dallas/Fort Worth, could be investigated further to understand specific reasons and implement measures to improve efficiency.
Further Analysis:

It would be beneficial to analyze delay data across all months and for multiple years to discern patterns or trends.
Comparing delay data with other metrics like the number of flights, passenger volume, or airport capacity could provide deeper insights.
A more in-depth analysis of specific airlines could reveal which carriers are more prone to delays and why.
